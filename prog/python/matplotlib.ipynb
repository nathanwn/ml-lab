```python
import numpy as np
from matplotlib import pyplot as plt
```

# 0. Show Plots in Jupyter Notebook

To show plots in Jupyter Notebook, run the following command:

```python
%matplotlib inline
```

The alternative is to run `plt.show()`.

# 1. Functional Matplotlib

## 1.1. Simple Line Plot

```python
x = np.linspace(start=0, stop=10, num=40)
y = x ** 2
plt.plot(x, y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

## 1.2. Multiple plots on one canvas

```python
plt.subplot(2, 2, 1) # 2 rows, 2 columns, plot number
plt.plot(x, y, 'red')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'blue')
plt.subplot(2, 2, 3)
plt.plot(y, x, 'green')
plt.subplot(2, 2, 4)
plt.plot(y, x, 'purple')
```

# 2. OO Matplotlib

## 2.1. Plot with manual subplots

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
axes2 = fig.add_axes([0.3, 0.5, 0.2, 0.3])
axes2.plot(y, x)
```

## 2.2. Grid of subplots

```python
fig, axes = plt.subplots(nrows=3, ncols=2)

for row in axes:
    for ax in row:
        ax.plot(x, y)

plt.tight_layout() # prevent overlapping
```

# 3. Figure Size

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 8))

for row in axes:
    for ax in row:
        ax.plot(x, y)

plt.tight_layout() # prevent overlapping
```

# 5. Style Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#6080AA', linewidth=3, alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#6080AA', linewidth=1, alpha=0.8, marker='o', markersize=5)
```
